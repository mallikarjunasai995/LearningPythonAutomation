```python
unsorted_data = pd.read_csv('Advance_Pandas/datasets/echocardiogram.csv')
unsorted_data = unsorted_data[:15]
print(unsorted_data)
```

```python
"""Defining our own indexes"""
unsorted_data.index = [4, 1, 2, 5, 3, 8, 9, 7, 6, 12, 10, 11, 9, 13, 14]
unsorted_data
```

```python
sorted_data = unsorted_data.sort_index()

sorted_data
```

```python
"""sorting in descending order"""
sorted_data = sorted_data.sort_index( ascending = False)

sorted_data
```

```python
"""by default , sorting will happen with row index .. what if we want the same with column index ?"""
unsorted_data.sort_index( axis = 1)
```

```python
"""can sort a particular column also (it acts like series in this stage)"""
sorted_data['epss'].sort_index()
```

```python
"""so far we have sorted the dataframe by their row and columns, now we are going to sort by column values in the cells"""
unsorted_data.sort_values(by = 'age')
```

```python
"""multiple column values sorting , below example data first sorted by the survival and then age"""
unsorted_data[['survival',
               'age',
               'epss',
               'lvdd',
               'wallmotion-score']].sort_values( by = ['survival',
                                                       'age'])
```
