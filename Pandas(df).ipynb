```python
"""series can create excel like sheet where index(heading)and rows(data). Two types of series can be created
1. through list
2. through dictionary, below example"""
country_capitals_dict = {'Afganistan': 'kabul',
                        'India': 'New Delhi',
                        'Bangladesh': 'Dhaka',
                        'China': 'Beijing',
                        'Italy': 'Rome'}
country_capitals_series = pd.Series(country_capitals_dict)
country_capitals_series
```

"""How do we access data from the series object"""
"""In Python, a multi-index, also known as a hierarchical index, is used to represent and manage data that has multiple dimensions or levels of indexing

"""


```python
country_capitals_series.values
country_capitals_series.index
country_capitals_series[2]
country_capitals_series['Afganistan','China']
```

```python
country_capitals_series[['Afganistan','China']]
```

```python
"""DataFrames are like excel sheet where index and labelled axes will be there. Dataframe can contain another data frame
for example , In any one cell we can have entire pandas dataframe.  
data frames are tabular representation of data"""
empty_dataframe = pd.DataFrame()
print(empty_dataframe)
```

```python
"""From list to insert the pandas dataframes"""
political_figures=["nelson","gandhi"]
df_from_list = pd.DataFrame(political_figures)
df_from_list
```

```python
"""From tuples"""
celebrities_tuple =('Robin','chiranjeevi')
df_from_tuple= pd.DataFrame(celebrities_tuple)
df_from_tuple
```

```python
"tuple created from list and pass it to dataframes"
countries=['USA','India']
political_figures=['Nelson','Gandhi']
politicians_tuples= list(zip(countries,political_figures))
politicians_tuples
```

```python
df_from_listoftuples = pd.DataFrame(politicians_tuples,columns=['Country','Political Names'])
df_from_listoftuples
```

```python
"""From Dictionary , creating pandas dataframes tabular data"""
countries=['USA','India']
political_figures=['Nelson','Gandhi']
politicians_dict = {'Politician_name': pd.Series(political_figures),
                   'Country':pd.Series(countries)}
politicians_dfdict = pd.DataFrame(politicians_dict)
politicians_dfdict
```

```python
#load data into dataframes
politicians_dfdict.to_csv('datasets/politicians.csv',index = False)
politicans_df_fromfile = pd.read_csv('datasets/politicians.csv')
politicans_df_fromfile
#type(politicans_df_fromfile)
```
